# Question 1 - Boolean Comparison

*   Give an example of boolean comparison with 'and'
*   Give an example of boolean comparison with 'or'




```
def can_run_for_president(age, is_natural_born_citizen):
    return is_natural_born_citizen and (age >= 35)

def student(age,school): return (age <= 20) or school


print("Can a 19-year-old run for president?", can_run_for_president(19,True)) 
print("I am Students?", student(18,False))
```

*   Create an if-else statement that will print 'Evening' if the clock is later than 16.00 and the sky color is 'Yellow'. 'Noon' if the clock is equal to or less than 16.00 and the sky color is 'Blue'. 'Morning' if the clock is less than 12.00 and the sky color is 'Orange'



```
Sky = 'Blue'
Clock = 08.00

if Clock > 16.00 and Sky == 'Yellow':
  print ('Evening')
elif Clock <= 16.00 and Sky == 'Blue':
  print ('Noon')
elif Clock < 12.00 and Sky == 'Orange':
  print ('Morning')
```

# Question 5 - Simple Function

Create a function to get the average of a list object

```
obj_list = [13.4, 12.5, 16.0, 5.5, 18.0]
def mean_list(inp_list):
    sum_num = 0
    for t in inp_list:
        sum_num = sum_num + t        
    avg = sum_num / len(inp_list)
    return avg
  
print("The average is", mean_list(obj_list))
```
