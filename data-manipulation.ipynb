```
import re #Import Regular Expression
text = 'Malming hari ini kami meracik cuka menjadi minuman yang bermanfaat'

a = re.findall(r'\w+ing',text) #Search by pattern ing
b = re.findall(r'\w*c\w*',text) #Search by pattern c
c = re.findall(r'\w*an\w*',text) #Search by pattern an

print("Kalimat berakhiran ing adalah ",", ".join(a))
print("Kalimat mengandung huruf C adalah ",", ".join(b))
print("Kalimat mengandung huruf an ",", ".join(c))
```

```
text = '''If you have never had a body fall out of a tree and knock you over, let me tell you what surprising thing that is. I have had nuts fall out of a tree and conk my head. Leaves have fallen on me, and twigs, and a branch during a storm. Bird slop, of course, everyone gets that. But a body? That is not your usual thing dropping out of a tree. It was a boy, close about my age, maybe twelve. Shaggy hair the color of dry dirt. Brown pants. Blue T-shirt. Bare feet. Dead. Didn't recognize him. My first thought was, Is this my fault? I bet this is my fault. Nula once had a knack for being around when trouble happened. She had not been around other kids much, though, and maybe did not know that most kids had a knack for being around when trouble happened.'''

class Counter():

    def __init__(self,teks):
        self.teks = teks

    def kata_berbeda(self):
        kata = Counter.kata(self)
        counts = {}
        for r in kata:
            if r in counts:
                counts[r] += 1
            else:
                counts[r] = 1
        return counts
    
    def kata(self):
        return self.teks.split()
        
def main():
    print('Total semua kata : ')
    print(Counter(text).kata())
    print('\nJumlah keseluruhan kata : ',len(Counter(text).kata()))
    print(Counter(text).kata_berbeda())
    print('\nJumlah Kata yang berbeda : ', len(Counter(text).kata_berbeda()))

main()
```
